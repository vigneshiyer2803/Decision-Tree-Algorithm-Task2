# Splitting of the dataset

```python
X=df.drop(['Species'], axis=1)
y=df['Species']
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3, random_state = 1)

print("Shape of X_train : ", X_train.shape, "\nShape of y_train : ", y_train.shape)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT= DecisionTreeClassifier(criterion='gini')
DT.fit(X_train, y_train)
```

```python
x=DT.score(X_test,y_test)
print('The accuracy is :', (x*100).round(2))
```

# Plotting of the Tree

```python
from sklearn import tree
rcParams['figure.figsize'] = 10,15
tree.plot_tree(DT,precision=5,rounded=True,filled=True);
#Plotting the decision tree
```
